```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#dataset load
df = pd.read_csv("Traffic_Violations.csv")
df
```

```python
# Checking number of rows and columns
df.shape
```

```python
# Showing all column names
df.columns
```

```python
# Checking data types of each column
df.dtypes
```

```python
# Display complete dataset information
# It shows total entries, non-null values, and datatype
df.info()
```

```python
# Checking null/missing values in each column
df.isnull().sum()
```

```python
# Removing duplicate rows from dataset
df = df.drop_duplicates()

# Checking shape after removing duplicates
df.shape
```

```python
# Selecting all numeric columns
num_cols = df.select_dtypes(include=np.number).columns

# Filling missing numeric values with median
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
# Selecting all categorical columns
cat_cols = df.select_dtypes(include="object").columns

# Filling missing categorical values with "Unknown"
for col in cat_cols:
    df[col] = df[col].fillna("Unknown")
```

```python
#  statistical summary of dataset
df.describe()
```

```python
# Checking if all missing values are handled
df.isnull().sum()
```

```python
# Using iloc to access rows by integer position
df.iloc[0:10]
```

```python
# Using loc to access rows by labels
df.loc[0:10]
```

```python
# Counting occurrences of each vehicle type
df['VehicleType'].value_counts()
```

```python
# Grouping data by state and counting total violations
df.groupby('State').size()
```

```python
# Selecting numeric columns only
numeric_df = df.select_dtypes(include=np.number)

# Finding correlation between numeric columns
numeric_df.corr()
```

```python
# Plotting histogram for all numeric columns
numeric_df.hist(figsize=(15,10))

# Display the graph
plt.show()
```

```python
# Plotting heatmap for correlation matrix
plt.figure(figsize=(12,8))
sns.heatmap(numeric_df.corr(), annot=False)

# Display heatmap
plt.show()
```

```python
# Plotting bar chart of top 10 most common vehicle types
df['VehicleType'].value_counts().head(10).plot(kind='bar')

# Adding title
plt.title("Top 10 Vehicle Types")

# Display graph
plt.show()
```

```python
# Pie chart showing top 5 vehicle types
vehicle_counts = df['VehicleType'].value_counts().head(5)

plt.figure(figsize=(8,8))
plt.pie(vehicle_counts, labels=vehicle_counts.index, autopct='%1.1f%%')

plt.title("Top 5 Vehicle Types Distribution")
plt.show()
```

```python
# Scatter plot for violation locations
plt.figure(figsize=(10,6))
plt.scatter(df['Latitude'], df['Longitude'])

plt.title("Violation Locations")
plt.xlabel("Latitude")
plt.ylabel("Longitude")

plt.show()
```

```python
# Saving cleaned dataset into new CSV file
df.to_csv("Traffic_Violations_Cleaned.csv", index=False)
```
